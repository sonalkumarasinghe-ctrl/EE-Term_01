```python
import numpy as np
import pandas as pd
```

```python
input_data = pd.read_csv('Housing.csv')
input_data.head()
```

```python
input_data.columns
```

```python
categorical_col1=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
input_data[categorical_col1]=input_data[categorical_col1].map(lambda x:1 if x=='yes' else 0)
```

```python
categorical_col2=['furnishingstatus']
input_data[categorical_col2]=input_data[categorical_col2].map(lambda x:2 if x=='furnished' else (1 if x=='semi-furnished' else 0))
input_data.head()
```

```python
y=input_data['price']
x=input_data[['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']]
print(x.shape,y.shape)
print(x)
```

```python
from sklearn.model_selection import train_test_split    
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
# X_train=np.reshape(X_train,(-1,1))
# X_test=np.reshape(X_train,(-1,1))
y_train,y_test=np.reshape(y_train,(-1,1)),np.reshape(y_test,(-1,1))
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
LinearRegression()
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
# Making Predictions
y_pred = regressor.predict(X_test)
results = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
results.head()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
MAE=mean_absolute_error(y_test, y_pred)
MSE=mean_squared_error(y_test, y_pred)
RMSE=np.sqrt(MSE)
print(f'MAE= {MAE}, MSE= {MSE}, RMSE= {RMSE}')
```

```python
regressor.score(X_test, y_test)
  #Error is considerably High
```
