```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
input_data=pd.read_csv('WineQT.csv')
input_data.head()
```

```python
quality_counts = input_data['quality'].value_counts().sort_index()
plt.figure(figsize=(8, 6))
sns.barplot(x=quality_counts.index, y=quality_counts.values, palette='viridis')
plt.title('Wine Quality Counts')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.set(style="darkgrid")
numerical_columns=input_data.select_dtypes(include=["int64","float64"]).columns
plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(input_data[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(input_data[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```
