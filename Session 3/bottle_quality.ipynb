```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
input_data=pd.read_csv('WineQT.csv')
input_data.head()
```

```python
quality_counts = input_data['quality'].value_counts().sort_index()
plt.figure(figsize=(8, 6))
sns.barplot(x=quality_counts.index, y=quality_counts.values, palette='viridis')
plt.title('Wine Quality Counts')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.set(style="darkgrid")
numerical_columns=input_data.select_dtypes(include=["int64","float64"]).columns
plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(input_data[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(input_data[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.swarmplot(x="quality", y="alcohol", data=input_data, palette='viridis')

plt.title('Swarm Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

```python
sns.set_palette("Pastel1")

plt.figure(figsize=(10, 6))

sns.pairplot(input_data)

plt.suptitle('Pair Plot for DataFrame')
plt.show()
```

```python
sns.boxplot(x='quality', y='alcohol', data=input_data)
```

```python
plt.figure(figsize=(15, 10))

sns.heatmap(input_data.corr(), annot=True, fmt='.2f', cmap='Pastel2', linewidths=2)

plt.title('Correlation Heatmap')
plt.show()
```

```python
#Creating a Linear Regression Model
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = 100 * np.random.rand(50, 1)
y = 3 * X + 10 + np.random.randn(50, 1) * 10
```

```python
# plot the data points
plt.scatter(X, y, color='blue', label='Data Points')
```

```python
model=LinearRegression()
model.fit(X, y)
Y_pred=model.predict(X)
```

```python
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
